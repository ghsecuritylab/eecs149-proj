```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import wave
import sys
import scipy as sp
import scipy.signal

spf = wave.open('AKWF_oboe_0001.wav','r')
#Extract Raw Audio from Wav File
#If Stereo
if spf.getnchannels() == 2:
    print('Just mono files')
    sys.exit(0)
signal = spf.readframes(-1)
signal = np.fromstring(signal, 'Int16')

plt.figure(1)
plt.title('Original Waveform')
plt.plot(signal)
plt.show()
```

```python
def upsample(original_signal, rate):
    result = []
    for _ in range(rate):
        for s in signal[::rate]:
            result.append(s)
    return result
```

```python
def sum_signals(signals):
    result = []
    for set in zip(*signals):
        result.append(sum(set))
    return result
```

```python
def display_for_c(signal):
    result = []
    result.append('{')
    for x in signal[:-1]:
        result.append(str(x))
        result.append(', ')
    result.append(str(signal[-1]))
    result.append('};')
    print(''.join(result))
```

```python
def string_for_c(signal):
    result = []
    result.append('{')
    for x in signal[:-1]:
        result.append(str(x))
        result.append(', ')
    result.append(str(signal[-1]))
    result.append('};')
    return ''.join(result)
```

```python
def scale_amplitude(signal, scalar):
    return [int(x * scalar) for x in signal]
```

```python
def num_samples_required(clk_f, target_f):
    return int(clk_f/target_f + 0.5)
```
